```python
#Setup the notebook

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("setup complete")
```

```python
#Load the file 
Penguin_filepath='C:/Users/SAGARISHA/Downloads/archive/penguins.csv'

penguins_data=pd.read_csv(Penguin_filepath,index_col='species',parse_dates=True)

```

```python
#Examine the data

penguins_data.head()
```

```python
penguins_data.tail()
```

```python
# Check the shape of the dataframe

print("The shape of the dataframe is:", penguins_data.shape)
```

```python
# Check for missing values
print("Missing values: \n", penguins_data.isnull().sum())
```

```python
# Remove duplicate rows

penguins_data.drop_duplicates(inplace=True)
```

```python

# Check for duplicate rows again
print(penguins_data.duplicated().sum())
```

```python

# Reindex the dataframe

penguins_data.reset_index(drop=True, inplace=True)
```

```python
# Check the shape of the dataframe after removing duplicates and reindexing
print("The shape of the dataframe is:", penguins_data.shape)
```

```python
# Get a summary of the numerical columns

print(penguins_data.describe())
```

```python
# Plot histograms of the numerical columns
penguins_data.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

The histograms for bill length, bill depth, body mass, and flipper length show a variety of distributions, with some being symmetrical and others being skewed. The bill length and bill depth data appear to be normally distributed, with peaks at 47 mm and 17 mm, respectively. The body mass data is slightly positively skewed, with a peak at 3700 g. The flipper length data is also positively skewed, with a peak at 200 mm.


In terms of center, the mean and median values are similar for bill length and bill depth, indicating a symmetrical distribution. For body mass and flipper length, the mean values are slightly higher than the median values, indicating a positive skew in the data.

The spread of the data is moderate to large, with the majority of values falling between the 25th and 75th percentiles. There are no significant outliers in any of the four variables.

In conclusion, the distributions of the bill length, bill depth, body mass, and flipper length data in penguins show a range of shapes, with some being symmetrical and others being skewed. The center and spread of the data also vary, but no significant outliers are present in any of the four variables."

```python
# Plot pairplot to visualize the relationships between the variables
sns.pairplot(penguins_data)
plt.show()
```

Summary:
"The pair plot of the bill length, bill depth, body mass, and flipper length data shows a range of relationships between the variables. The histograms for each variable show a variety of distributions, with some being symmetrical and others being skewed.

The scatter plots between bill length and body mass, as well as body mass and flipper length, show positive correlations. This indicates that as one variable increases, the other variable is likely to increase as well. On the other hand, the scatter plots between bill length and bill depth, as well as bill depth and flipper length, show weaker or no correlations.

In terms of center and spread, the majority of the data falls within the interquartile range for each variable, with no significant outliers present.

In conclusion, the pair plot provides a visual representation of the relationships between the bill length, bill depth, body mass, and flipper length variables in penguins. The histograms show a variety of distributions, and the scatter plots reveal positive correlations between some of the variables. The center and spread of the data is moderate, with no significant outliers present."





```python
# Plot boxplots of the numerical columns to detect outliers
penguins_data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(20,15))
plt.show()
```

Summary:

The box represents the interquartile range (IQR), which contains the middle 50% of the data, while the whiskers represent the range of the data.

In terms of center, the median values of bill length, bill depth, body mass, and flipper length are close to each other, indicating a moderate spread of the data.

In terms of spread, the IQR values for bill length, bill depth, body mass, and flipper length are moderate, with no significant outliers present.

In conclusion, the box plot provides a visual representation of the center, spread, and distribution of the bill length, bill depth, body mass, and flipper length variables in penguins. The median values are close to each other, indicating a moderate spread of the data, and the IQR values are moderate, with no significant outliers present."



```python
# Plot a heatmap to show the correlation between the variables
sns.heatmap(penguins_data.corr(), annot=True)
plt.show()
```

Summary:
"The heat map of the bill length, bill depth, body mass, and flipper length data shows a visual representation of the relationships between the variables. The color intensity of each cell in the matrix indicates the strength of the correlation between two variables.

The heat map shows that bill length and flipper length have a moderate positive correlation, indicating that as one variable increases, the other is likely to increase as well. Body mass and flipper length also have a strong positive correlation. On the other hand, the correlations between bill length and bill depth, as well as bill depth and flipper length, are negatively correlated.

In conclusion, the heat map provides a visual representation of the relationships between the bill length, bill depth, body mass, and flipper length variables in penguins. The heat map shows strong positive correlations between flipper length and body mass, as well as bill depth and flipper length, while the correlations between bill length and bill depth, as well as body mass and flipper length, are weaker."
